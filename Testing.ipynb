## Preprocessing Research Categories

The original dataset is [Fields of Research (FoR) Classification](https://figshare.com/articles/dataset/Fields_of_Research_FoR_Classification/10305443?file=35317927) by Megan Hardeman which originates from the [2008 ANZSRC FoR classification](https://www.arc.gov.au/manage-your-grant/classification-codes-rfcd-seo-and-anzsic-codes) 

```python
import pandas as pd
import numpy as np
df = pd.read_csv("ResearchCategories.csv")
df = df.drop(["Unnamed: 0", "Unnamed: 4", "Unnamed: 5"], axis=1)
```

```python
last_discipline = ""
for i in range(len(df)):
    curr = df['Discipline'][i]
    if type(curr) == str and not curr.isnumeric():
        last_discipline = curr
    else:
        df.loc[i, 'Discipline'] = last_discipline
```

```python
last_field = ""
for i in range(len(df)):
    curr = df['Field'][i]
    if type(curr) == str and not curr.isnumeric():
        last_field = curr
    else:
        df.loc[i, 'Field'] = last_field
```

```python
for i in range(len(df)):
    curr = df.loc[i, 'Topic']
    if type(curr) == float and np.isnan(curr):
        df.drop(i, axis=0, inplace=True)
```

```python
df.reset_index(inplace=True, drop=True)
display(df)
```

```python
for i in range(len(df)):
    df.loc[i, 'Discipline'] = df.loc[i, 'Discipline'].title()
```

```python
for i in range(len(df)):
    if "not elsewhere classified" in df.loc[i, 'Topic']:
        df.loc[i, 'Topic'] = df.loc[i, 'Topic'][:-24]
```

```python
df.to_csv("ResearchCategories.csv")
```
